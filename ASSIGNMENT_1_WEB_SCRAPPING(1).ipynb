##  Write a python program to display IMDB’s Top rated 100 movies’ data

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://www.imdb.com/chart/top/')

page
soup = BeautifulSoup(page.content)

soup

years = []

for i in soup.find_all('span',class_="secondaryInfo"):
     years.append (i.text)
        
years[0:100]

titles = []

for i in soup.find_all('td',class_="titleColumn"):
    titles.append(i.text.replace('\n',''))
    
titles[0:100]

imdb_ratings = []

for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    imdb_ratings.append(i.text.replace('\n',''))
    
imdb_ratings[0:100]

movies = pd.DataFrame({'Movies':titles[0:100],
                       'Years':years[0:100],
                       'Imdb_rating':imdb_ratings[0:100],})


movies
```

## Write a python program to scrape mentioned news details

```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.cnbc.com/world/?region=world')
page
soup = BeautifulSoup(page.content)
soup

time = []

for i in soup.find_all('span',class_= "LatestNews-wrapper")[0:20]:

    time.append (i.text)
    
time
```

```python
title = []

for i in soup.find_all('a',class_= "LatestNews-headline")[0:20]:

    title.append (i.text)
    
title
```

```python

link = []
for i in soup.find_all('a',href = True)[100:120]:
  link.append(i['href'])
link
  
```

```python
pd.DataFrame({'title':title,'time':time,'link':link})
```

## Write a python program to scrape mentioned details from dineout.co.in :

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = BeautifulSoup(page.content)
soup.prettify
```

```python
restaurant_name =[]
for i in soup.find_all('div',class_='restnt-info cursor'):
    restaurant_name.append(i.a.text)
restaurant_name
```

```python
cuisine = []
for i in soup.find_all('div',class_='detail-info'):
    cuisine.append(i.text.replace('|',''))
cuisine
```

```python
cuisine1 = []
for i in cuisine:
    cuisine1.append(i[22:])

cuisine1
```

```python
location =[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
location
```

```python
rating =[]
for i in soup.find_all('div',class_='restnt-rating rating-4'):
    rating.append(i.text)
rating
```

```python
print(len(rating))
```

```python
image = []
for i in soup.find_all('img',class_='no-img'):
    image.append(i['data-src'])
image
```

```python
Dineout=pd.DataFrame({'Restuarant_name':restaurant_name,'Cuisine':cuisine1,'Location':location,'Rating':rating,'Image_URL':image})
Dineout
```
